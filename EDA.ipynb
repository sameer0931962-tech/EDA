## Question 2: Explain overfitting and underfitting in ML. How can you detect and prevent them?
## Hint: Discuss bias-variance tradeoff, cross-validation, and regularization techniques

## **1. Overfitting**

**Meaning:**
Overfitting happens when a model learns the training data **too closely**, including noise and small details.
So it performs **very well on training data** but **poorly on test data**.

**Why it happens:**

* Model is too complex
* Too many parameters
* Not enough training data

**How to detect:**

* Training accuracy is high but test/validation accuracy is low
* Large difference between training and validation error

**How to prevent:**

* Use **cross-validation** to check performance
* Apply **regularization** (L1, L2)
* Use **dropout** in neural networks
* Use a **simpler model**
* Collect **more data**
* Use **early stopping**

---

## **2. Underfitting**

**Meaning:**
Underfitting happens when a model is **too simple** and cannot learn the patterns in data.
It performs **poorly on both training and test data**.

**Why it happens:**

* Model not complex enough
* Not enough training
* Missing important features

**How to detect:**

* Low accuracy on both training and test
* High bias error

**How to prevent:**

* Use a **more complex model**
* Add **more features**
* Train for **longer**
* Reduce strong regularization

---

## **Bias–Variance Connection (Short)**

* **Overfitting = low bias + high variance**
* **Underfitting = high bias + low variance**
  Goal: find balance using cross-validation and regularization.

## Question 3:How would you handle missing values in a dataset? Explain at least three methods with examples.
## Hint: Consider deletion, mean/median imputation, and predictive modeling.

# **Handling Missing Values in a Dataset**

When a dataset has empty or missing values, we must clean them before training a model.
Here are **three common methods**:

---

## **1. Deletion Method**

### **(a) Remove Rows**

If only a few rows have missing values, we can delete those rows.

**Example:**
Dataset:

| Age | Salary |
| --- | ------ |
| 25  | 30,000 |
| —   | 45,000 |
| 30  | 50,000 |

The second row has missing Age → delete it.

### **When to use:**

* When missing data is **very small** (less than 5%).
* When deleting won’t lose important information.

---

## **2. Mean / Median Imputation**

Replace missing values with **mean**, **median**, or **mode** (for categorical data).

**Example (Mean):**
Ages: 20, 25, — , 30
Mean = (20 + 25 + 30) / 3 = 25
Missing value becomes **25**.

**Example (Median):**
Values: 5, —, 50
Median = 5
Missing value becomes **5**.

### **When to use:**

* When missing data is moderate.
* When the column is **numerical**.

---

## **3. Predictive Modeling (Advanced Method)**

Use another machine-learning model to **predict the missing value** based on other features.

**Example:**
Dataset has missing “Salary”, but we know:

* Age: 28
* Experience: 3 years
* Education: Graduate

Train a small ML model (like Regression or Decision Tree) to predict the salary for missing rows.

### **When to use:**

* When many values are missing.
* When data is important and cannot be deleted.
* When more accuracy is needed.

---

# **Summary Table**

| Method                     | Description                             | Example                             | Use When                              |
| -------------------------- | --------------------------------------- | ----------------------------------- | ------------------------------------- |
| **Deletion**               | Remove rows/columns with missing values | Delete row with missing Age         | Missing data is small                 |
| **Mean/Median Imputation** | Fill missing values with statistics     | Fill missing Age with mean          | Numeric data                          |
| **Predictive Modeling**    | Build model to predict missing values   | Predict Salary using other features | Many values missing, require accuracy |


## Question 4:What is an imbalanced dataset? Describe two techniques to handle it (theoretical + practical).
## Hint: Discuss SMOTE, Random Under/Oversampling, and class weights in models.

# **Imbalanced Dataset**

An **imbalanced dataset** is when one class (category) has **many more samples** than the other.
This causes the model to **favor the majority class** and perform poorly on the minority class.

**Example:**
Fraud detection:

* 990 normal transactions
* 10 fraud transactions
  Model may predict everything as “normal” and still get 99% accuracy.

---

# **Techniques to Handle Imbalanced Data**

## **1. Oversampling (Example: SMOTE)**

### **Theory:**

Oversampling increases the number of minority class samples.
**SMOTE (Synthetic Minority Oversampling Technique)** creates **new synthetic (not duplicate)** samples by combining existing minority samples.

### **Practical Example:**

If the dataset has:

* 100 majority class
* 10 minority class

SMOTE can generate 90 synthetic minority samples → balance becomes 100–100.

**Tools:**

* Python: `from imblearn.over_sampling import SMOTE`
* ML models become more fair and accurate.

---

## **2. Undersampling (Random Undersampling)**

### **Theory:**

Undersampling reduces the number of majority class samples to match the minority class.
This keeps the dataset balanced but **may remove useful data**.

### **Practical Example:**

If you have:

* 1000 majority
* 100 minority

Random undersampling may reduce majority class to 100 → new dataset = 100–100.

**Use when:**

* Dataset is very large
* Losing some majority samples is okay

---

## **3. Class Weights (Alternative Method)**

### **Theory:**

Instead of balancing data, you tell the algorithm to **give more importance (weight)** to the minority class.

### **Practical Example:**

In logistic regression or decision trees:

* Majority weight = 1
* Minority weight = 5

This forces the model to pay more attention to minority samples.

**Tools:**

* In sklearn: `class_weight='balanced'`

---

# **Short Summary**

| Concept                  | Explanation                            | Example                                |
| ------------------------ | -------------------------------------- | -------------------------------------- |
| **Imbalanced Dataset**   | One class has far fewer samples        | Fraud detection (10 fraud, 990 normal) |
| **SMOTE (Oversampling)** | Creates synthetic minority samples     | Increase fraud cases from 10 → 100     |
| **Undersampling**        | Removes majority samples               | Reduce normal cases 1000 → 100         |
| **Class Weights**        | Give more importance to minority class | Weight fraud class higher              |

## Question 5:  Why is feature scaling important in ML? Compare Min-Max scaling andStandardization.
## Hint: Explain impact on distance-based algorithms (e.g., KNN, SVM) and gradien descent.


# **Why Feature Scaling is Important in ML**

Feature scaling means bringing all numerical features to a similar range.
It is important because many ML algorithms are **sensitive to the scale of data**.

### **Why it matters:**

1. **Distance-based algorithms** (KNN, K-Means, SVM)

   * These algorithms calculate distance between points.
   * If one feature has a large range (e.g., salary 0–1,00,000) and another has a small range (e.g., age 20–50),
     the large-range feature dominates the distance → wrong results.

2. **Gradient Descent (used in Linear Regression, Logistic Regression, Neural Networks)**

   * Large-scale values cause **slow convergence** and unstable training.
   * Scaling speeds up learning and improves accuracy.

**In short:**
Feature scaling ensures **fair contribution of all features** and improves **model performance and training speed**.

---

# **Two Main Scaling Techniques**

## **1. Min–Max Scaling (Normalization)**

### **Formula:**

[
X' = \frac{X - X_{min}}{X_{max} - X_{min}}
]

### **Range:**

0 to 1

### **Use When:**

* Data does **not have many outliers**
* Needed in deep learning, neural networks
* Works well with KNN, K-Means, SVM

### **Example:**

Age values: 20, 30, 40
Min = 20, Max = 40
30 → (30−20)/(40−20) = 0.5

---

## **2. Standardization (Z-score Scaling)**

### **Formula:**

[
X' = \frac{X - \mu}{\sigma}
]

### **Range:**

Mean = 0, Standard deviation = 1
(Not fixed range)

### **Use When:**

* Data has **outliers**
* Works well with Logistic Regression, Linear Regression, SVM, Neural Networks

### **Example:**

Value = 50, Mean = 40, Std = 5
Z = (50−40)/5 = 2

---

# **Short Comparison Table**

| Feature               | Min-Max Scaling               | Standardization                                |
| --------------------- | ----------------------------- | ---------------------------------------------- |
| Formula               | (X−min)/(max−min)             | (X−mean)/std                                   |
| Range                 | 0 to 1                        | Mean 0, std 1                                  |
| Sensitive to Outliers | Yes                           | No                                             |
| Best for              | Neural networks, KNN, K-Means | Regression, SVM, models using gradient descent |

## Hint: Consider categorical variables with ordinal vs. nominal relationships.


# **Label Encoding vs One-Hot Encoding**

## **1. Label Encoding**

**What it does:**

* Converts categories into **numbers** (0, 1, 2, 3…).
* Example:
  Color → {Red = 0, Blue = 1, Green = 2}

**When to use:**

* When the categories have a **natural order (ordinal data)**.
  Examples:

  * Size: Small < Medium < Large
  * Education: Primary < Secondary < Graduate

**Advantage:**

* Simple and keeps dataset small.

**Disadvantage:**

* Adds **false ranking** if used for unordered categories.
  (e.g., Red > Blue is meaningless)

---

## **2. One-Hot Encoding**

**What it does:**

* Creates **separate columns** for each category with 0/1 values.
* Example:
  Color →

  * Red   = [1, 0, 0]
  * Blue  = [0, 1, 0]
  * Green = [0, 0, 1]

**When to use:**

* When categories have **no order (nominal data)**.
  Examples:

  * Colors (Red, Blue, Green)
  * Cities (Delhi, Mumbai, Chennai)
  * Gender (Male, Female)

**Advantage:**

* No false ranking; helpful for most ML algorithms.

**Disadvantage:**

* Increases number of columns (high dimensionality).

---

# **Short Comparison Table**

| Feature  | Label Encoding      | One-Hot Encoding          |
| -------- | ------------------- | ------------------------- |
| Output   | Numbers (0,1,2…)    | Binary columns (0/1)      |
| Best For | **Ordinal** data    | **Nominal** data          |
| Problem  | Creates false order | Many columns              |
| Example  | Size: S=0, M=1, L=2 | Colors → Red, Blue, Green |

---

# **When to Prefer What?**

* **Use Label Encoding** → When categories have **meaningful order**.
* **Use One-Hot Encoding** → When categories have **no order**, and you want to avoid ranking errors.

# **Question 7: Google Play Store Dataset – Category vs Rating (Easy Explanation + Python Code)**

### **a) Relationship Between App Categories and Ratings (Easy Explanation)**

Different app categories receive different average ratings because users have different expectations from each type of app.

### **Categories with Highest Ratings (usually):**

* **Health & Fitness**
* **Education**
* **Productivity**
* **Lifestyle**

**Reason:**
These apps usually solve specific problems, work simply, and users rate them well if they provide value. These categories also have fewer low-quality apps.

### **Categories with Lowest Ratings (usually):**

* **Games**
* **Tools**
* **Social**

**Reason:**
These categories have **too many apps**, including many low-quality or buggy apps.

### **Conclusion:**

* Categories with **focused purpose** → higher ratings
* Categories with **many apps or high competition** → lower ratings

---

### **What this code does:**

* Loads the dataset
* Removes missing rating values
* Groups apps by category
* Calculates average rating of each category
* Prints highest & lowest rated categories


## Python Code (Simple and Clean)

## You can paste this directly into Jupyter/Colab.

```python
import pandas as pd

# Load dataset
df = pd.read_csv("googleplaystore.csv")

# Clean data: keep only valid ratings
df = df[df['Rating'].notna()]
df['Rating'] = pd.to_numeric(df['Rating'], errors='coerce')
df = df.dropna(subset=['Rating'])

# Calculate average rating per category
category_ratings = df.groupby('Category')['Rating'].mean().sort_values(ascending=False)

print("Top Categories (Highest Average Rating):")
print(category_ratings.head(5))

print("\nLowest Categories (Lowest Average Rating):")
print(category_ratings.tail(5))
```

# **Question 8: Titanic Dataset (Easy Explanation + Python Code)**

---

# **(a) Survival Rate by Passenger Class (Pclass)**

### **Result (Easy Explanation):**

* **1st Class → Highest survival rate**
* **2nd Class → Medium survival rate**
* **3rd Class → Lowest survival rate**

### **Why did 1st Class survive more?**

1. **First class passengers were closer to lifeboats**
   Their cabins were on upper decks, so they reached lifeboats quickly.

2. **Crew prioritized wealthy passengers**
   Social status influenced rescue decisions.

3. **Better access to information**
   First class passengers were notified earlier than lower classes.

So, **1st Class = safest**, **3rd Class = most affected**.

---

# **(b) Did Children Survive More Than Adults?**

### **Result (Easy Explanation):**

Yes.
**Children (Age < 18)** had a **higher survival rate** than adults.

### **Why?**

1. **The rule “Women and children first”**
   Ship crew rescued children before adults.

2. **Children often stayed with mothers (women had higher survival too)**
   This increased their chance of getting seats on lifeboats.

**Conclusion:**
✔ Children survived more
✖ Adults had lower survival rate

```python
import pandas as pd

# Load Titanic dataset
df = pd.read_csv("titanic.csv")   # adjust path if needed

# Drop missing values in Age and Survived for safe calculation
df = df.dropna(subset=['Survived', 'Age', 'Pclass'])

# -----------------------------------------
# Part (a): Survival Rate by Pclass
# -----------------------------------------

survival_by_class = df.groupby('Pclass')['Survived'].mean() * 100
print("Survival Rate by Passenger Class (%):")
print(survival_by_class)

# -----------------------------------------
# Part (b): Survival Rate by Age Group
# -----------------------------------------

# Create age groups
df['AgeGroup'] = df['Age'].apply(lambda x: 'Child' if x < 18 else 'Adult')

survival_by_age = df.groupby('AgeGroup')['Survived'].mean() * 100
print("\nSurvival Rate by Age Group (%):")
print(survival_by_age)
```

# ✅ **Question 9 — Flight Price Prediction Dataset**

## **(a) How do flight prices vary with days left until departure?**

### **Short Theory (Easy Explanation)**

* When **days left are many** (20–30 days), prices are **low**.
* As you get **closer to travel day**, prices **increase gradually**.
* In the **last 3–5 days**, prices **jump suddenly (exponential surge)** because demand is high and seats are fewer.
* Best booking time = **15–25 days before travel** (average lowest price).

---

## **(b) Compare prices across airlines (Example: Delhi → Mumbai route)**

### **Short Theory (Easy Explanation)**

* Some airlines keep **low base fare** → consistently cheaper.
* Premium airlines have **better service + more demand** → higher fares.
* For the same route:

  * **Low-cost airlines** (IndiGo, SpiceJet) → cheaper.
  * **Full-service airlines** (Vistara, Air India) → premium.

---

```python

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset directly from GitHub
url = "https://raw.githubusercontent.com/MasteriNeuron/datasets/main/flight_price.csv"
df = pd.read_csv(url)

# Preview dataset
print("Dataset Preview:\n", df.head())

# -------------------------------
# Part (a) - Price vs Days Left
# -------------------------------
plt.figure(figsize=(8,5))
sns.scatterplot(x="days_left", y="price", data=df, alpha=0.5)
sns.lineplot(
    x="days_left",
    y="price",
    data=df.groupby("days_left")["price"].mean().reset_index(),
    color="red"
)
plt.title("Flight Price vs Days Left Until Departure")
plt.xlabel("Days Left")
plt.ylabel("Price")
plt.show()

# Average price by days left
avg_price = df.groupby("days_left")["price"].mean().reset_index()
print("\nAverage Price by Days Left:\n", avg_price.head(20))

```

```python
delhi_mumbai = df[(df["source_city"] == "Delhi") & (df["destination_city"] == "Mumbai")]
airline_avg = delhi_mumbai.groupby("airline")["price"].mean().sort_values()

print("\nAverage Prices for Delhi-Mumbai Route:\n", airline_avg)

plt.figure(figsize=(8,5))
sns.barplot(x=airline_avg.index, y=airline_avg.values)
plt.title("Average Airline Prices (Delhi-Mumbai)")
plt.ylabel("Average Price")
plt.xticks(rotation=45)
plt.show()
```
